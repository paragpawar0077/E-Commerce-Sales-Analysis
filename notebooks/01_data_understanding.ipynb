```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("../data/olist_orders_dataset.csv")
```

```python
customers = pd.read_csv("../data/olist_customers_dataset.csv")
```

```python
order_items = pd.read_csv("../data/olist_order_items_dataset.csv")
```

```python
payments = pd.read_csv("../data/olist_order_payments_dataset.csv")
```

```python
reviews = pd.read_csv("../data/olist_order_reviews_dataset.csv")
```

```python
products = pd.read_csv("../data/olist_products_dataset.csv")
```

```python
sellers = pd.read_csv("../data/olist_sellers_dataset.csv")
```

```python
category_translation = pd.read_csv("../data/product_category_name_translation.csv")
```

```python
for name, df in [("orders", orders), ("customers", customers), 
                  ("order_items", order_items), ("payments", payments),
                  ("reviews", reviews), ("products", products), 
                  ("sellers", sellers)]:
    print(f"\n=== {name} ===")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
```

```python
print("=== NULLS ===")
for name, df in [("orders", orders), ("order_items", order_items), 
                  ("products", products), ("reviews", reviews),("payments", payments),("sellers", sellers),("customers", customers)]:
    print(f"\n{name}:")
    null_counts = df.isnull().sum()[df.isnull().sum() > 0]
    if len(null_counts) == 0:
        print("No nulls found")
    else:
        print(null_counts)
```

```python
print("\n=== ORDER STATUS ===")
print(orders['order_status'].value_counts())
```

```python
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col], format='mixed')

print("Date columns converted successfully")
print(orders[date_columns].dtypes)
print(orders['order_purchase_timestamp'].head())
```
